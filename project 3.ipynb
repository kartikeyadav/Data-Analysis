# total application

```python
TA=df["Application id"].nunique()
```

# approved application

```python
AP=(df["tdecision"]=="Approve").sum()
```

# booked application

```python
BA=(df["Booking"]=="Y").sum()
```

# approval rate

```python
AR=AP/TA
```

# booking rate

```python
BR=BA/AP
```

# new booking amount

```python
NBA=df.Booking_Amt.sum()
```

```python
metrics=pd.DataFrame({"total application":TA,"approved application":AP,"booked application":BA,"approval rate":AR,"booking rate":BR,"new booking amount":NBA},index=[0])
```
